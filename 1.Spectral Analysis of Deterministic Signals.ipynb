```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import inv
from numpy.fft import fft, ifft


X = np.array([10, 5, 6, 6, 2, 4, 3, 4, 5, 0, 0, 0, 0])
N = len(X)
k = np.arange(N) 
A = 10  

tmpmu = 2-1/2  
tmpmu = 1 
mu = np.arange(N)
K = np.outer(k, mu) 
W = np.exp(+1j * 2*np.pi/N * K)

x_test = 1/N * np.matmul(W, X)

plt.stem(k, np.real(x_test), label='real',
             markerfmt='C0o', basefmt='C0:', linefmt='C0:')
plt.stem(k, np.imag(x_test), label='imag',
             markerfmt='C1o', basefmt='C1:', linefmt='C1:')    

plt.plot(k, np.real(x_test), 'C0o-', lw=0.5)
plt.plot(k, np.imag(x_test), 'C1o-', lw=0.5)
plt.xlabel(r'sample $k$')
plt.ylabel(r'$x[k]$')
plt.legend()
plt.grid(True)

print("Verification with numpy ifft:", np.allclose(np.fft.ifft(X), x_test))
print("DC Component:", np.mean(x_test))

print("Matrix K:")
print(K)
print("\nMatrix W:")
print(W)
```
